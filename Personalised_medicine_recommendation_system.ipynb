# Reason Based Recommendation System

```python
med
```

```python
med_rating = med.groupby("drugName").count()['usefulCount'].reset_index()
med_rating.rename(columns={'usefulCount' : 'drug_rating'},inplace=True)
med_rating
```

```python
avg_rating = med.groupby("drugName").mean()['rating'].reset_index()
avg_rating.rename(columns={'rating' : 'avg_rating'},inplace=True)
avg_rating
```

```python
medi = med_rating.merge(avg_rating, on='drugName')
medi
```

```python
medicine = medi[medi['drug_rating']>=1].sort_values("avg_rating", ascending=False)
```

```python
medicine_df = medicine.merge(med, on='drugName').drop_duplicates("drugName")[["drugName", "condition", "drug_rating", "avg_rating", "review"]]
```

```python
medicine_df
```

```python
medicine_df.groupby('drugName').count()['drug_rating']
```

```python
pt = medicine_df.pivot_table(index='drugName', columns = 'condition', values='drug_rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend(drugName):
  index = np.where(pt.index==drugName)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1], reverse=True)[1:6]

  data = []
  for i in similar_items:
    item = []
    temp_df = med[med['drugName'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('drugName')['drugName'].values))
    item.extend(list(temp_df.drop_duplicates('drugName')['condition'].values))
    item.extend(list(temp_df.drop_duplicates('drugName')['review'].values))
    item.extend(list(temp_df.drop_duplicates('drugName')['rating'].values))

    data.append(item)


  return data

```

```python
recommend('femhrt')
```
